```python
print(reduced_y.Department.unique())
print(reduced_X.Hospital_type_code.unique())
print(reduced_X.Hospital_region_code.unique())
print(reduced_X['Severity of Illness'].unique())
print(reduced_X['Type of Admission'].unique())
print(reduced_X['Age'].unique())
Ordinalencoder_cols = ['Severity of Illness','Type of Admission','Age']
OnehotEncoder_cols = ['Department','Hospital_type_code','Hospital_region_code']
# one hot encoding for Department , hospital type code , hospital region code
# ordinal encoding for severity of illness , Type of Admission
```

```python
#  now checking any missing value in data
```

```python
cols_with_missing_values = []
for cols in reduced_X.columns:
    if reduced_X[cols].isnull().any():
        cols_with_missing_values.append(cols)
cols_with_missing_values
```

```python
# filling missing values with city code of hospital
for i in range(0,len(reduced_X.City_Code_Hospital)):
    if pd.isnull(reduced_X['City_Code_Patient'][i]):
        reduced_X['City_Code_Patient'][i] = reduced_X['City_Code_Hospital'][i]

for i in range(0,len(reduced_y.City_Code_Hospital)):
    if pd.isnull(reduced_y['City_Code_Patient'][i]):
        reduced_y['City_Code_Patient'][i] = reduced_y['City_Code_Hospital'][i]
```

```python
X_train = reduced_X.drop('Stay', axis = 1)
stay = ['Stay']
y_train = reduced_X[stay]
```

```python
import numpy as np
arr = np.array(y_train)
```

Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
# copy
label_X_train = X_train.copy()  # training data features
label_X_valid = reduced_y.copy() # testing data
label_y_train = arr.copy() # training data output 
# apply ordinal encoder
ordinal_encoder = OrdinalEncoder()
label_X_train[Ordinalencoder_cols] = ordinal_encoder.fit_transform(X_train[Ordinalencoder_cols])
label_X_valid[Ordinalencoder_cols] = ordinal_encoder.transform(reduced_y[Ordinalencoder_cols])
label_y_train = ordinal_encoder.fit_transform(arr)
```

Encoding Severity of illness

moderate-> 0

minor -> 1

extreme -> 2

Encoding Type of Admission

Emergency -> 0

trauma -> 1

urgent -> 2

One hot encoding 

```python
from sklearn.preprocessing import OneHotEncoder

OH_encoder = OneHotEncoder(handle_unknown='ignore',sparse = False)
OH_cols_train = pd.DataFrame(OH_encoder.fit_transform(X_train[OnehotEncoder_cols]))
OH_cols_valid = pd.DataFrame(OH_encoder.transform(reduced_y[OnehotEncoder_cols]))

# put back indexes
OH_cols_train.index = X_train.index
OH_cols_valid.index = reduced_y.index

# remove categorical columns
num_X_train = label_X_train.drop(OnehotEncoder_cols,axis =1)
num_X_valid  = label_X_valid.drop(OnehotEncoder_cols,axis = 1)

# adding columns 
Final_X_train = pd.concat([num_X_train,OH_cols_train],axis =1)
Final_X_valid = pd.concat([num_X_valid , OH_cols_valid], axis = 1)
```

```python
from sklearn.ensemble import RandomForestRegressor

Health_model = RandomForestRegressor(random_state=1)
Health_model.fit(Final_X_train,label_y_train.ravel())
preds = Health_model.predict(Final_X_valid)
print(preds)
```

```python
import math 
for i in range(0,len(preds)):
    preds[i] = math.ceil(preds[i])
prediction = pd.Series(preds)
```
